```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.

import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(seed=1234)  # replace with YOUR last 4 digits

sensor_a = rng.normal(loc=25.0, scale=3.0, size=200)
sensor_b = rng.normal(loc=27.0, scale=4.5, size=200)
timestamps = rng.uniform(low=0.0, high=10.0, size=200)
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Color Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a tnotiitle, and a legend.
plt.figure(figsize=(10, 6))
plt.scatter(timestamps, sensor_a, s=22, alpha=0.7, label='Sensor A')
plt.scatter(timestamps, sensor_b, s=22, alpha=0.7, label='Sensor B')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (C)')
plt.title('Sensor Temperatures vs Time')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Overlaid histogram of Sensor A and Sensor B temperature distributions.
# Use 30 bins, alpha=0.5 for transparency so both distributions are visible.
# Add vertical dashed lines at each sensor's mean.
# Include a legend labeling each sensor.
mean_a = np.mean(sensor_a)
mean_b = np.mean(sensor_b)

plt.figure(figsize=(10, 6))
plt.hist(sensor_a, bins=30, alpha=0.5, label='Sensor A', color='tab:blue')
plt.hist(sensor_b, bins=30, alpha=0.5, label='Sensor B', color='tab:orange')

plt.axvline(mean_a, color='tab:blue', linestyle='--', linewidth=2, label=f'Sensor A Mean: {mean_a:.2f} C')
plt.axvline(mean_b, color='tab:orange', linestyle='--', linewidth=2, label=f'Sensor B Mean: {mean_b:.2f} C')

plt.xlabel('Temperature (C)')
plt.ylabel('Count')
plt.title('Temperature Distribution: Sensor A vs Sensor B')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
# Side-by-side box plot comparing Sensor A and Sensor B distributions.
# Label x-axis with sensor names, y-axis with "Temperature (deg C)".
# Add a horizontal dashed line at the overall mean of both sensors combined.
overall_mean = np.mean(np.concatenate([sensor_a, sensor_b]))

plt.figure(figsize=(8, 6))
plt.boxplot([sensor_a, sensor_b], labels=['Sensor A', 'Sensor B'])
plt.ylabel('Temperature (deg C)')
plt.title('Sensor Temperature Distributions (Box Plot)')
plt.axhline(overall_mean, color='gray', linestyle='--', linewidth=2, label=f'Overall Mean: {overall_mean:.2f} deg C')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```
